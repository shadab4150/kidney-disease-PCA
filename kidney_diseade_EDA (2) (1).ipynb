```python
def clas(x):
  if x=='ckd\t':
    return 'ckd'
  else:
    return x
```

```python
kidney.classification=kidney.classification.apply(clas,convert_dtype=True)
```

```python
kidney.classification.unique()
```

```python
kidney.wc=pd.to_numeric(kidney.wc,errors='coerce')
kidney.rc=pd.to_numeric(kidney.rc,errors='coerce')
```

```python
def clas(x):
  if x=='\tno':
    return 'no'
  elif x=='yes':
    return 'yes'
  elif x=='no':
    return 'no'
  else:
    return 'unknown'
```
